```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#criando o dataframe sem  o cabeçalho.
df = pd.read_csv('./treino_sinais_vitais_qPA_versus_Gravidade.csv', header=None)
```

```python
df.head(5)
```

```python
#no csv temos 3 colunas, e de acordo com a descrição da tarefa no moodle seriam:
#1 - Indice
#2 - Entrada
#3 - Saída
```

```python
#porém ao criarmos o DF o python entende que a primeira coluna faz parte dos dados
#ajustando ficaria assim:
df = df[[1,2]]
df.head(5)
```

```python
#colocando cabeçalho
df.columns = ['Entrada', 'Saida']
df.head(5)
```

```python
#com os dados "ajustados" podemos plotar os pontos
df.plot(x='Entrada',y='Saida')
```

```python
#nao ficou muito boa a visualização, vamos tentar com pontos
df.plot(x='Entrada',y='Saida', kind='scatter')
#agora os dados parecem mais organizados.
```

### Regressão Linear

```python
# Como toda RL, precisamos de uma variavel dependente 
#e de uma ou mais independentes, vamos definir abaixo:
#extraindo a primeira coluna do df, convertendo en um array bidimensional, armazena na variável X sendo variavel Dependente
X = df.iloc[:, 0].values.reshape(-1, 1)
y = df.iloc[:, 1].values  #iloc[:, 1] ":"todas as linhas da coluna 2 "1" começa em 0.
```

```python
#utilizando a função de regressão linear scikit
rl = LinearRegression()

#treinando o modelo e neste passo é ajustado o coeficiente e o intercepto para 
#essa funçao utiliza o método dos mínimos quadrados
#com isto, podemos gerar previsões, após o treinamento.
rl.fit(X, y)

# Gerando previsões, recebe a matriz bidimencional X criada acima para podermos avaliar o desempenho do modelo
y_pred = rl.predict(X)

# Imprimindo os coeficientes da regressão
print("Coeficiente: ", rl.coef_)
print("Intercepto: ", rl.intercept_)

#plotando a regressão

plt.scatter(X, y, color='gray', label='Dados CSV')
plt.plot(X, y_pred, color='red', label='Regressão Linear')
plt.xlabel('Entrada')
plt.ylabel('Saída')
plt.legend()
plt.show()
```
